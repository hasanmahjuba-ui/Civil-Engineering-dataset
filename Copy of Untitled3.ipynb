```python
# 1) Run this cell to upload from your computer (click Choose Files)
from google.colab import files
uploaded = files.upload()  # pick the CSV file you downloaded from Kaggle

# 2) Load & inspect with pandas
import io, pandas as pd
# replace 'yourfile.csv' with the actual filename if needed:
fname = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[fname]))

# quick inspections
print("Filename:", fname)
print("Shape:", df.shape)
display(df.head(10))      # first 10 rows
print("\nInfo:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nBasic stats:")
display(df.describe(include='all').transpose())
```
